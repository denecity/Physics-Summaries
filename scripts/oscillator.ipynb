```python
import numpy as np
import plotly.express as px

T = np.pi

a_0 = 2

def f_initial(t):
    return a_0 * (1 - np.cos(t))
    
    
def f_later(t, T):
    return a_0 * (np.cos(t - T) - np.cos(t))

x = np.linspace(0, 20, 1000)

fx = np.zeros_like(x)

for i, t in enumerate(x):
    if t < T:
        fx[i] = f_initial(t)
    else:
        fx[i] = f_later(t, T)
        
fig = px.line(x=x, y=fx, labels={'x': 'Time', 'y': 'Displacement'}, title='Displacement of the Oscillator')
fig.add_vline(x=T, line_dash="dash", line_color="red")
fig.show()
```
